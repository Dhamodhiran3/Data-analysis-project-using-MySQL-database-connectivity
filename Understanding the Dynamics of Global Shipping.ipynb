```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Count the frequency of each unique value in the 'Mode_of_Shipment' column
shipment_mode_counts = df['Mode_of_Shipment'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(shipment_mode_counts, labels=shipment_mode_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Shipment Modes')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Create a stock chart
plt.figure(figsize=(10, 6))
plt.plot(df['Cost_of_the_Product'], label='Cost of Product', color='blue')
plt.plot(df['Discount_offered'], label='Discount Offered', color='orange')
plt.fill_between(df.index, df['Cost_of_the_Product'], df['Discount_offered'], where=(df['Cost_of_the_Product'] > df['Discount_offered']), interpolate=True, color='green', alpha=0.3)
plt.fill_between(df.index, df['Cost_of_the_Product'], df['Discount_offered'], where=(df['Cost_of_the_Product'] < df['Discount_offered']), interpolate=True, color='red', alpha=0.3)
plt.title('Stock Chart: Cost of Product vs Discount Offered')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Count the frequency of each unique value in the 'Mode_of_Shipment' column
shipment_mode_counts = df['Mode_of_Shipment'].value_counts()

# Plot the histogram
plt.figure(figsize=(8, 6))
shipment_mode_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Distribution of Shipment Modes')
plt.xlabel('Shipment Modes')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
